```python
# 1 Scrape the details of most viewed videos on YouTube from Wikipedia. Url
# = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:

driver=webdriver.Chrome()
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
RANK=[]
name=[]
artist=[]
views=[]
date=[]
```

```python
table=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]')
rows=table.find_elements(By.TAG_NAME,'tr')

for row in rows[1:]:
    cells=row.find_elements(By.TAG_NAME,'td')
    if len(cells)>=4:
        rank=cells[0].text
        vedio_name=cells[1].text
        uploader=cells[2].text
        views=cells[3].text
        date=cells[4].text
        
        print(f'{rank},{vedio_name},{uploader},{views},{date}')
```
